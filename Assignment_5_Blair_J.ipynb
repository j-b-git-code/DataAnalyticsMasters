```python
#Block 1 - Data input and setup Q1-2
# Import the appropriate Python libraries.
import pandas as pan
import seaborn as sea

# Load the CSV file into a dataframe.
df_realestate = pan.read_csv('http://bit.ly/real-estate-data')

#Check the data

#Getting info
# Question 1 - How many records (rows) are in the data? 414 (could use .shape)
# Question 2 - What is the datatype for the convienience store column? int64
df_realestate.info()


```

```python
#Block 2 - Adjusting the Field Names and Questions 3-4

#Changing the Column names - format is '<current name>':'<new name>'
df_realestate = df_realestate.rename(columns = {'No': 'sale_number',
                                                'X1 transaction date':'sale_date',
                                                'X2 house age':'house_age',
                                                'X3 distance to the nearest MRT station':'distance_to_station',
                                                'X4 number of convenience stores':'num_convenience_stores',
                                                'X5 latitude':'latitude',
                                                'X6 longitude':'longitude',
                                                'Y house price of unit area':'home_price_per_unit'})

#check the result
df_realestate.info()

# Question 3 - preview the first 7 records.
#              what is the age of house on the last? 34.5 years
df_realestate.head(7)

#Question 4 - generate descriptive stats for the numeric columns
#             What is the highest value in the price column?  117.50
df_realestate.describe()
```

```python
#Block 3 - Generate the Histogram and Question 5

#generating histogram of the housing price column
sea.displot(df_realestate, x='home_price_per_unit')

#Question 5 - Housing price tends to center around 60? No, centers around 40
```

```python
#Block 4 - Generate a Scatterplot and Question 6

#Generating Scatterplot of relationship between age and price
sea.scatterplot(x= 'house_age', y='home_price_per_unit',data=df_realestate)

#Question 6
#It appears there is a strong positive relationship betwen age and price?
#No, there appears to be no strong relationship between age and price
```

```python
#Block 5 - Generate a Correlation Matrix and Questions 7-11

#Generating Correlation Matrix
df_realestate.corr()

#Question 7
#What is the correlation coeff for distance to station and home price? -0.674

#Question 8
#It appears that the farther from the train the higher the house price?
#False, the inverse is true.

#Question 9
#What is the correlation coeff for num of convenience stores and price? 0.571

#Question 10
#Based on the relationship, as num of convenience stores increases so does price
#True, there is a corelation between those items

#Question 11
#Based on the coefficients above it is clear that a change in home age directly
#causes a change in home price.
#False, corelation is not causation, also there are many counter examples
```

```python
#Block 6 - Generate a heatmap

#preparing data - omiting the sale number, sale date longitude, and lattitude
#iloc says we want all rows but ommit the above columns
# corr() generate the correlation
corr = df_realestate.iloc[:,[2,3,4,7]].corr()
#Generate the seaborn heatmap
sea.heatmap(corr, annot=True)
```
